```python
inventory={"1":{"Name":"pepsi","Quantity":20,"Price":80},                      
           "2":{"Name":"bourbon","Quantity":50,"Price":25},
           "3":{"Name":"apple","Quantity":60,"Price":25},
           "4":{"Name":"lizol","Quantity":95,"Price":95},
           "5":{"Name":"britannia_cake","Quantity":30,"Price":30},
           "6":{"Name":"lays","Quantity":50,"Price":20},
           "7":{"Name":"amul_chocolate","Quantity":50,"Price":110}}
```

```python
# for display  
def inventory_display():
    print("%10s %20s %10s %20s"%("S.no","Item","Quantity","Cost/Item(in Rupee)"))
    for x,y in inventory.items():
        print("%10s %20s %10s %10d" %(x,y["Name"],y["Quantity"],y["Price"]))
```

```python
#price of the items in inventory
price_of_item=[]
for outerkey in inventory:
    price_of_item.append(inventory[outerkey]["Price"])

#quantity of items in inventory
quantity_in_stock=[]
for outerkey in inventory:
    quantity_in_stock.append(inventory[outerkey]["Quantity"])

#name of  items in inventory
name_item=[]
for outerkey in inventory:
    name_item.append(inventory[outerkey]["Name"])
```

```python
name_ordered=[]              # name of items ordered
quantity_ordered=[] 
stock=[]
inp="yes"

# to take the input from the customer for which items he need
def item_needed(inp):
    while (inp=='yes'):
        input1=input("Enter the item you want to purchase : ")
        print("\n")
        stock.append(input1)
        input2=int(input("how much quantity is needed : "))
        print("\n")
        b=stock_check()                                   #give the value b which the quantity of item ordered in the stock
        if input2>b:
            print("item out of stock")
        else:
            name_ordered.append(input1)
            quantity_ordered.append(input2)
            stock_update()                                       #updates the item quantity remaining in the stock
        inp=input("Do u want to add more items(yes/no): ")
        print("\n")
        
    return name_ordered,quantity_ordered



```

```python
# customer name and address

def name_address():
    address=0
    name="x"
    name=input("Please Enter your name : " )
    print("\n")
    address=int(input("how much is the distance of ur home from this store : "))
    print("\n")
    return name,address

# for calculating the delivery charges
def delivery_charges(address):
    if address<5:
        charges=10
    elif address>=5 and address<10:
        charges=12
    elif address>=10 and address<15:
        charges=14
    else:
        charges=20
    return charges

#price of ordered items by the user
def price_of_ordered_items():
    for i in range(len(name_ordered)):
        x=name_item.index(name_ordered[i])
        y=price_of_item[x]
        ordered_item_price.append(y)
        return ordered_item_price
    
#items cost without delivery charges
def items_cost():
    list_of_items_after_multiplying_with_quantity=list(map(lambda a,b:a*b,quantity_ordered,ordered_item_price))
    defitems_cost=sum(map(lambda a,b:a*b,quantity_ordered,ordered_item_price))
    return defitems_cost,list_of_items_after_multiplying_with_quantity    

# Total cost
def total_cost(charges,defitems_cost):
    total_price=charges + defitems_cost
    return total_price
    
#stock check function to check if stock is available
def stock_check():
    x=1
    i=0
    while i<x:
        a=name_item.index(stock[i])
        b=quantity_in_stock[a]
        return b
# for updating the stocks
quantity_in_stock1=quantity_in_stock.copy()
def stock_update():
    for i in range(len(quantity_ordered)):
        a=name_item.index(name_ordered[i])
        quantity_in_stock[a]=quantity_in_stock1[a]-quantity_ordered[i]
        return quantity_in_stock1,quantity_in_stock
    
## to generate the bill
def generate_bill():
    space=' '
    print(space*20,"Your Own Store")
    print(space*20,"______________")
    print("\n")
    print(space*25,"BILL")
    print("\n")
    print("Customer's Name:",name)
    print("Distance from the store:",address,"Km")
    print("\n")
    space=' '
    print(space*20,"List of Items Purchased")
    print(space*20,"_______________________")
    print("\n")
    print("%10s %15s %25s %15s" %("Name","Quantity","Cost/Item(in Rupees)","Price"))
    for i in range(0,len(name_ordered)):
        print("%10s %15d %25d %15d" %(name_ordered[i],quantity_ordered[i],
        ordered_item_price[i],list_of_items_after_multiplying_with_quantity[i]))
    print("\n")
    print("%58s %5d %5s" %("Item Price:",defitems_cost,"Rupees"))
    print("%58s %5d %5s" %("Delivery Charges:",charges,"Rupees"))
    print("%58s %5d %5s" %("Grand Total:",total_price,"Rupees"))
    print("\n")
    print(space*15,"THANK YOU FOR SHOPPING")
    print(space*19,"VISIT US AGAIN")
```

```python
## Main Program for the Shopping
ordered_item_price=[]
name_ordered=[]            
quantity_ordered=[]
inp='yes'
n='yes'
while(n=='yes'):
    print("these are all the items we have in the store")
    print("\n")
    inventory_display()                                  #to display the list of items available for shopping
    print("\n")
    name_ordered,quantity_ordered=item_needed(inp)         # to take the input from the user
    price_of_ordered_items()
    n=input('yes to shop more and no to proceed to checkout: ')
    print("\n")
ordered_item_price = price_of_ordered_items()                          # to get the cost/item price of the ordered items
name,address = name_address()                                           # to take input for customer details
charges = delivery_charges(address)                                   # to calculate the delivery charges 
defitems_cost,list_of_items_after_multiplying_with_quantity = items_cost()    # to calculate cost of the items
total_price = total_cost(charges,defitems_cost)                               # to calculate the total price
print("\n")
generate_bill()                                                                # to generate the bill

```
